```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes = load_diabetes()
```

```python
X = diabetes.data[:, np.newaxis, 2]
y = diabetes.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error:", round(mse, 2))
print("Coefficient of determination:", round(r2, 2))
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred)
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.title("Linear Regression: BMI vs Disease Progression")
plt.show()
```

Conclusion:

En este ejercicio se aplicó un modelo de regresión lineal para analizar la progresión de la diabetes en función del índice de masa corporal (IMC). El coeficiente de determinación obtenido indica que el IMC explica una parte importante, pero no total, de la variabilidad en la progresión de la enfermedad. Esto sugiere que, si bien el IMC es un factor relevante, existen otras variables clínicas que influyen en la evolución de la diabetes.
