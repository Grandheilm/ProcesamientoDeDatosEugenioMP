```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Cargar dataset diamonds desde seaborn
diamonds = sns.load_dataset("diamonds")

# Mostrar primeras filas
diamonds.head()
```

```python
# Verificar valores nulos
diamonds.isna().sum()
```

```python
X = diamonds[['carat']]
y = diamonds['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error:", round(mse, 2))
print("Coefficient of determination:", round(r2, 2))
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred)
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Linear Regression: Carat vs Price")
plt.show()
```

```python
X2 = diamonds[['carat', 'depth']]
y = diamonds['price']
```

```python
X2_train, X2_test, y_train, y_test = train_test_split(
    X2, y, test_size=0.3, random_state=42
)
```

```python
model2 = LinearRegression()
model2.fit(X2_train, y_train)

y_pred2 = model2.predict(X2_test)
```

```python
mse2 = mean_squared_error(y_test, y_pred2)
r22 = r2_score(y_test, y_pred2)

print("Mean squared error:", round(mse2, 2))
print("Coefficient of determination:", round(r22, 2))
```
