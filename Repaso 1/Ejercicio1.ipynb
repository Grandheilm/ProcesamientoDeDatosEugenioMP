```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ingreso = [
1452,1361,1426,1470,1456,1430,1354,1442,1394,1459,
1399,1458,1537,1425,1445,1439,1348,1450,1431,1446,
1485,1405,1461,1490,1426
]

habitaciones = [
23,47,21,39,37,29,23,44,45,16,
30,42,54,27,34,15,19,38,44,47,
43,38,51,61,39
]
```

```python
df = pd.DataFrame({
    "Habitaciones ocupadas": habitaciones,
    "Ingreso del restaurante": ingreso
})
```

```python
plt.scatter(df["Habitaciones ocupadas"], df["Ingreso del restaurante"])
plt.xlabel("Habitaciones ocupadas")
plt.ylabel("Ingreso del restaurante")
plt.title("Gráfica de correlación lineal simple")
plt.show()
```

```python
r = np.corrcoef(habitaciones, ingreso)[0,1]
r
```

La gráfica de dispersión muestra una tendencia creciente entre el número de habitaciones ocupadas y el ingreso diario del restaurante. A medida que aumenta la cantidad de habitaciones ocupadas, el ingreso del restaurante tiende a incrementarse. Sin embargo, los puntos no se encuentran perfectamente alineados, lo que indica que la relación no es fuerte.

De acuerdo con la gráfica, existe una correlación lineal positiva moderada entre las variables analizadas. Esto sugiere que el ingreso del restaurante está influenciado por el número de habitaciones ocupadas, aunque también intervienen otros factores.

El coeficiente de correlación de Pearson obtenido es:

**r = 0.423**

Este valor confirma que existe una correlación positiva moderada entre el número de habitaciones ocupadas y el ingreso diario del restaurante, validando la interpretación realizada a partir de la gráfica.
